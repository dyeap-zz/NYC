```python
def discretizeBorough(borough,numOfBoroughs):  
    column_encoder = preprocessing.LabelEncoder()
    borough_encoded = column_encoder.fit_transform(borough)
    newBoroughFeatures = [[0 for x in range(numOfBoroughs)] for x in range(len(borough))] 
    v = 0
    for v in range(numOfBoroughs):
        for i in range(len(borough)):
            if(borough_encoded[i] == v):
                newBoroughFeatures[i][v] = 1
            else:
                newBoroughFeatures[i][v] = 0
    return np.array(newBoroughFeatures)

    print borough_encoded
borough = cnypd['BOROUGH']
boroghFeatureSplit = discretizeBorough(borough,5)
```

```python
def encodeVehicleType(vehicleCodeOne):
    vehicle_encoder1 = preprocessing.LabelEncoder()
    vehicle_en1 = vehicle_encoder1.fit_transform(vehicleCodeOne)
#Ambulance = 0
#bicycle=1
#bus=2
#firetruck = 3
#largcomveh = 4
#livery vehicle = 5
#motorcycle = 6
#other = 7
#Passenger = 8
#pedicab = 9
#pickuptruck = 10
#scooter = 11
#small comveh = 12
#sport utility = 13
#taxi = 14
#unknown = 15
#van = 16
    #5 categories of vehicle
    #government vehicle = ambulance/fire truck/bus/taxi[0,3,2,14]
    #unknown vehicles = unknown/OTHER/livery vehicles[15,7,5]
    #generic vehicles = passenger vehicle/STATION WAGON/van/SMALL COM VEH(4 TIRES)[8,13,16,12]
    #large vehicles = LARGE COM VEH(6 OR MORE TIRES)/PICK-UP TRUCK[4,10]
    #two wheeler = bicycle/motorcycle/scooter/pedicab[1,6,11,9]
    numOfCategories = 5
    numOfInstances = len(vehicle_en1)
    vehicleFeature = [[0 for x in range(numOfCategories)] for x in range(numOfInstances)] 
    for i in range(numOfInstances):
        if(vehicle_en1[i] == 0 or vehicle_en1[i] == 3 or vehicle_en1[i] == 2 or vehicle_en1[i]==14 ):
            vehicleFeature[i][0] = 1
        elif(vehicle_en1[i] == 15 or vehicle_en1[i] == 7 or vehicle_en1[i] == 5):
            vehicleFeature[i][1] = 1
        elif(vehicle_en1[i] == 8 or vehicle_en1[i] == 13 or vehicle_en1[i] == 16 or vehicle_en1[i] ==12 ):
            vehicleFeature[i][2] = 1
        elif(vehicle_en1[i] == 4 or vehicle_en1[i] == 10 ):
            vehicleFeature[i][3] = 1
        elif(vehicle_en1[i] == 1 or vehicle_en1[i] == 6 or vehicle_en1[i] == 11 or vehicle_en1[i]==9 ):
            vehicleFeature[i][4] = 1
        else:
            print "A VEHICLE TYPE WAS NOT ENCODED"
            break
    return vehicleFeature
vehicleCodeOne= cnypd['VEHICLE_TYPE_CODE_1']
vehicleFeature = encodeVehicleType(vehicleCodeOne)
    
```

```python
def dateEncode(dateCol):
    #Separte into different months
    dateCol = np.array(dateCol)
    numOfMonths = 12
    numOfInstances = len(dateCol)
    dateFeature = [[0 for x in range(numOfMonths)] for x in range(numOfInstances)] 
    for i in range(numOfInstances):
        if(dateCol[i] == '1'):
            dateFeature[i][0] = 1
        elif(dateCol[i] == '2'):
            dateFeature[i][1] = 1
        elif(dateCol[i] == '3'):
            dateFeature[i][2] = 1
        elif(dateCol[i] == '4'):
            dateFeature[i][3] = 1
        elif(dateCol[i] == '5'):
            dateFeature[i][4] = 1
        elif(dateCol[i] == '6'):
            dateFeature[i][5] = 1
        elif(dateCol[i] == '7'):
            dateFeature[i][6] = 1
        elif(dateCol[i] == '8'):
            dateFeature[i][7] = 1
        elif(dateCol[i] == '9'):
            dateFeature[i][8] = 1
        elif(dateCol[i] == '10'):
            dateFeature[i][9] = 1
        elif(dateCol[i] == '11'):
            dateFeature[i][10] = 1
        elif(dateCol[i] == '12'):
            dateFeature[i][11] = 1
        else:
            print "A DATE TYPE WAS NOT ENCODED"
            print dateCol[i]
            break
    return dateFeature

```

```python
def timeEncode(timeCol):
    #morning, midday and night are the three division here
    timeCol = np.array(timeCol)
    numOfCategories = 3
    numOfInstances = len(timeCol)
    timeFeature = [[0 for x in range(numOfCategories)] for x in range(numOfInstances)] 
    for i in range(numOfInstances):
        if(timeCol[i] == '0' or timeCol[i] == '1' or timeCol[i] == '2' or timeCol[i]=='3' or timeCol[i] == '4' or timeCol[i] == '5' or timeCol[i] == '6' or timeCol[i] == '20' or timeCol[i] == '21' or timeCol[i] == '22' or timeCol[i]=='23' or timeCol[i] == '24' ):
            timeFeature[i][0] = 1
        elif(timeCol[i] == '7' or timeCol[i] == '8' or timeCol[i] == '9' or timeCol[i]=='10' or timeCol[i] == '11' or timeCol[i] == '12' ):
            timeFeature[i][1] = 1
        elif(timeCol[i] == '13' or timeCol[i] == '14' or timeCol[i] == '15' or timeCol[i]=='16' or timeCol[i] == '17' or timeCol[i] == '18' or timeCol[i] == '19' ):
            timeFeature[i][2] = 1
        else:
            print "A TIME TYPE WAS NOT ENCODED"
            break
    return timeFeature
    
time = cnypd['TIME'].str.split(":",1).str[0]
#print time

lol= timeEncode(time)
print pd.DataFrame(lol).shape
```
